# Installation Docker

## essayer

        sudo apt-get install docker.io

----------------------------------------------------------------
## ce que j'ai fait

https://docs.docker.com/engine/installation/linux/ubuntu/#install-using-the-repository

ouvrir un terminal et taper pout mettre à jour l'index package apt :

        sudo apt-get update

- instalation de Trusty 14.04 (permet à Docker d'utiliser les pilotes de stockages aufs):

        sudo apt-get install \
            linux-image-extra-$(uname -r) \
            linux-image-extra-virtual


- Installation de Docker

Docker CE :    (Docker EE -> pour entreprise)

    1. packets permettant l'utilisation d'un dépots sur HTTPS :
    
        sudo apt-get install \
            apt-transport-https \
            ca-certificates \
            curl \
            software-properties-common

    2. ajouter la clé Docker GPG officiel :

        curl -fsSL https://download.docker.com/linux/ubuntu/gpg | sudo apt-key add -

Vérifier que la clé fingerprint est : 9DC8 5822 9FC7 DD38 854A E2D8 8D81 803C 0EBF CD88.

        sudo apt-key fingerprint 0EBFCD88

    3. configuration d'un dépôt stable.

Remarque: la sous-commande `lsb_release -cs` donne le nom de votre distribution Ubuntu, tel que xenial.
amd64:

        sudo add-apt-repository \
            "deb [arch=amd64] https://download.docker.com/linux/ubuntu \
            $(lsb_release -cs) \
            stable"

    4. installation Docker

        sudo apt-get install docker-ce

## Groupe Docker

Ajouter un groupe Docker et mettre l'utilisateur à l'interieur, pour éviter de lancer Docker avec Sudo systématiquement

        sudo addgroup docker
        sudo usermod -aG docker nom-utilisateur
        sudo /etc/init.d/docker restart

# Utilisation Docker
----------------------------------------------------------------

* Affiche toutes les commandes de docker : **docker --help **

* Affiche les paramètres d'une commande : **docker [command] --help**

----------------------------------------------------------------

** Les principales commandes :**

----------------------------------------------------------------

* Cherche une image sur le dockerhub (que j'explique dans la partie suivante) : **docker search**
* Télécharge une image depuis le dockerhub : **docker pull**
* Envoie une image sur le dockerhub : **docker push**
* Liste les images disponibles : **docker images**
* Supprime une image : **docker rmi**
* Crée un conteneur : **docker run**
* Éteint un conteneur : **docker stop**
* Liste les conteneurs démarrés : **docker ps**, docker ps -a *pour voir tout les conteneurs*
* Affiche les processus en cours d'un conteneur : **docker top**
* Supprime un conteneur : **docker rm id_du_conteneur**, docker rm \`docker ps -a -q\`, ou, docker rm \$(docker ps -aq) *pour supprimer tout les conteneurs*
* Supprimer une image : **docker rmi id_ou_nom_de_l_image** docker rmi \$(docker images -q) *pour supprimer toutes les images*
* Crée un conteneur avec un dockerfile : **docker build**
* Sauvegarde un conteneur ou une image au format tar.gz : **docker save**
* Gère les réseaux : **docker network**
* Gère les volumes : **docker volume**

----------------------------------------------------------------

Avant de commencer, voici la liste des arguments que nous utiliserons dans cette partie :
* -t : Fournit un terminal au docker
* -i : Permet d'écrire dans le conteneur (couplé à -t)
* -d : Exécute le conteneur en arrière plan
* -v : Permet de monter un répertoire local sur le conteneur
* -p : Permet de binder un port sur le conteneur vers un port sur le host
* -e : Permet l'ajout d'une variable d'environnement
* --name : Donne un nom au conteneur
* --rm : Détruit le conteneur une fois terminé
* -w : Choisit le répertoire courant (dans le conteneur)
* --link : Permet de faire un lien entre deux conteneurs
