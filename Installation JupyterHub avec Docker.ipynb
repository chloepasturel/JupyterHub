# Installation Conteneur JupyterHub

- Créer un conteneur docker appelé jupyterhub,

ouvrir un terminal et taper :

    docker run -p 8000:8000 --name jupyterhub jupyterhub/jupyterhub jupyterhub


- aller dans le conteneur jupyterhub pour faire les config,

ouvrir un autre terminal et taper :

    docker exec -it jupyterhub bash

---------------------------------------------------------------------------

## Installation préalable

on se trouve maintenant dans le `conteneur`

- Installer notebook :

        pip install notebook

- Installer vim pour lire et modifier fichier :

        apt-get update
        apt-get install vim

- ajouter des utilisateurs avec mot de passe (tout compléter):

        adduser ploup

Dans un navigateur web aller à l’adresse : http://127.0.0.1:8000/

entrer les identifiants de l’utilisateur créé → se retrouve dans un notebook

---------------------------------------------------------------------------

## Créer un certificat SSL.

- Aller dans le fichier /srv du conteneur (on devrait normalement se situer dans le fichier /srv/jupyterhub) :

        cd /srv ou cd ..

- **créer fichier mycert.pem et mykey.key :**

        openssl req -x509 -nodes -days 365 -newkey rsa:1024 -keyout mykey.key -out mycert.pem

- vérifier si tout est là :

        ls
    
retourne -> jupyterhub  mycert.pem	mykey.key

---------------------------------------------------------------------------

## Configuration du conteneur

- Aller dans le fichier /srv/jupyterhub :

        cd /srv/jupyterhub ou cd jupyterhub
        
- Créer un fichier de configuration :

        jupyterhub --generate-config

jupyterhub_config.py est créer, on y trouve toutes les configurations que l'on peut faire

- Modifier le fichier :

        vim jupyterhub_config.py


Rappel de quelques commandes dans vim

- i – pour passer dans le mode insertion
- echap – pour sortir du mode insertion
- :q! – pour quitter sans enregistrer
- :wq – pour enregistrer

Modifier le fichier selon besoin et l’enregistrer (voir suite pour les modifications)

Pour que les modification soit prise en conte :
- sortir du conteneur :

        exit

- stopper le conteneur :

**docker stop jupyterhub**

- Puis le relancer :

**docker start jupyterhub**

normalement les modification sont pris en compte

pour avoir le défilement des logs en temps réel (**docker attach jupyterhub**)

---------------------------------------------------------------------------

## Modification du fichier de configuration


**Admin**

la définition d'un admin n'est pas obligatoire

- Définition des admins.

c.Authenticator.admin_users = {'nom_admin1', 'nom_admin2'}

- Autorisation d'accés aux serveurs des autres utilisateur aux admins

c.JupyterHub.admin_access = True


**Connection sécurisé SSL**

- certificat ssl

c.JupyterHub.ssl_cert = ‘/srv/mycert.pem’
    
- clé ssl

c.JupyterHub.ssl_key = ‘/srv/mykey.key’

** Changement de répertoire**

*cela permet aux utilisateur de naviguer à partir du fichier home (pour voir les fichiers des autres utilisateurs) à partir de leur notebook, mais de commencer toujours dans leur répertoire personnel.*

- Définition du répertoire où les utilisateurs arrive lorsque l’interface du notebook démarre (ici /home).

c.Spawner.notebook_dir = '/home'

- L'URL sur laquelle le serveur single-user doit commencer.

c.Spawner.default_url = '/tree/{username}'

*{username} sera remplacé par le nom de l'utilisateur.*

---------------------------------------------------------------------------

copier coller dans le fichier de configuration :

```
############################################################
####################### MODIFICATION #######################
############################################################   

# definition des admins
c.Authenticator.admin_users = {'ploup'}

# autorisation acces serveurs autre utilisateurs aux admins
c.JupyterHub.admin_access = True

# ssl
c.JupyterHub.ssl_cert = '/srv/mycert.pem'
c.JupyterHub.ssl_key = '/srv/mykey.key'

# repertoire
c.Spawner.notebook_dir = '/home'
c.Spawner.default_url = '/tree/{username}'

############################################################
```
